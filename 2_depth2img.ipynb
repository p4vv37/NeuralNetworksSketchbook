```python
import torch
import requests
from PIL import Image
from diffusers import StableDiffusionDepth2ImgPipeline

pipe = StableDiffusionDepth2ImgPipeline.from_pretrained(
   "stabilityai/stable-diffusion-2-depth",
   torch_dtype=torch.float16,
).to("cuda")
```

```python

init_img = Image.open("/home/pawel/git/NeuralNetworksSketchbook/car.jpg").convert("RGB")
i = 1024
generator = torch.Generator(device="cuda").manual_seed(i)
```

```python
i = 1037
generator = torch.Generator(device="cuda").manual_seed(i)
prompt = " A photo of a sport car, from the left, highly detailed, retro, unreal engine, masterpiece, trending on artstation, featured on pixiv, cinematic composition, beautiful lighting, sharp, details, highly detailed, uhd"
n_propmt = ""
imgs = pipe(prompt=prompt, image=init_img, negative_prompt=n_propmt, guidance_scale=9, strength=0.7, generator=generator).images

imgs[0]
```

```python
i += 1
generator = torch.Generator(device="cuda").manual_seed(1037)
prompt = " A photo of a gigantic truck, highly detailed, retro, unreal engine, masterpiece, trending on artstation, featured on pixiv, cinematic composition, beautiful lighting, sharp, details, highly detailed, uhd"
imgs = pipe(prompt=prompt, image=init_img, negative_prompt=n_propmt, guidance_scale=9, strength=0.7, generator=generator, num_inference_steps=50).images
imgs[0]
```
